## 3. Data Preparation

### 3.1 Drop Some Columns

```python
df_2 = df_1.drop(columns = ['Education', 'Income','AnyHealthcare','NoDocbcCost','CholCheck','HvyAlcoholConsump'])
df_2.head()
```

### 3.2 Feature Engineering

```python
# Only one hot encode Sex column because its a nominal value not binary value
new_sex = pd.get_dummies(df_2['Sex']).rename(columns={0.0: "Female", 1.0: "Male"}).astype('float64')
new_sex.head()
```

```python
df_3 = pd.concat([df_2, new_sex], axis = 1).drop(columns = 'Sex')
df_3.head()
```

### 3.3 Remove Duplicated

```python
print("duplicated rows:",df_3.duplicated().sum()) 
```

```python
df_3 = df_3.drop_duplicates(ignore_index = True)
print("duplicated rows:",df_3.duplicated().sum()) 
```

## 4. Modeling

### 4.1 Naive Bayes

```python
# Split dataset
X = df_3.drop(columns = 'HeartDiseaseorAttack')
y = df_3['HeartDiseaseorAttack']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,recall_score,f1_score, precision_score, confusion_matrix

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 150)

clf_nb = GaussianNB()
clf_nb.fit(X_train, y_train)
y_pred = clf_nb.predict(X_test)
```

```python
def eval_score(y_test, y_pred, model_name = "Model"):
    print(f"{model_name} Score")
    print(f"Accuracy : {accuracy_score(y_test, y_pred)}")
    print(f"Recall : {recall_score(y_test, y_pred)}")
    print(f"Precision : {precision_score(y_test, y_pred)}")
    print(f"F1-Score: {f1_score(y_test, y_pred)}")
```

```python
def eval_with_cm(y_test, y_pred, model_name= "Model"):
    sns.heatmap(pd.DataFrame(confusion_matrix(y_test, y_pred)), annot = True, fmt = 'g')
    plt.title(f"Confusion Matrix {model_name}")
    plt.ylabel("Actual Label")
    plt.xlabel("Predicted Label")
    plt.show()
```

```python
eval_score(y_test, y_pred, "Naive Bayes")
```

```python
eval_with_cm(y_test, y_pred, "Naive Bayes")
```

### 4.2 Decsion Tree

```python
from sklearn.tree import DecisionTreeClassifier

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 150)

clf_dct = DecisionTreeClassifier()
clf_dct.fit(X_train, y_train)
y_pred = clf_dct.predict(X_test)
eval_score(y_test, y_pred, "Decision Tree")
```

```python
eval_with_cm(y_test, y_pred, "Decision Tree")
```

### 4.3 Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 50)

clf_lr = LogisticRegression()
clf_lr.fit(X_train, y_train)
y_pred = clf_lr.predict(X_test)
eval_score(y_test, y_pred, "Logistic Regression")
```

```python
eval_with_cm(y_test, y_pred, "Logistic Regression")
```
